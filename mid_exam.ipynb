```python
# hiển thị dữ liệu

import seaborn as sns
plt.figure(figsize=(14,10))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm') # vẽ mã trận tương quan (có thể sử dụng để lựa chọn biến x đầu vào)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

y = dataset['price'].values # lấy y data là cột cuối cùng

# lựa chọn x có độ tương quan cao với y

x = dataset[['wheel-base','length','width','curb-weight','num-of-cylinders','engine-size','bore','horsepower']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.3)
```

```python
# xây dựng mô hình Linear

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
y_predict
```

```python
# tính toán độ chính xác

from sklearn.metrics import r2_score
Ac = r2_score(y_test, y_predict)
print('The accuracy of this model is R-square = ','{:.2f}'.format(Ac))
```

```python
from sklearn.metrics import mean_squared_error
import math
mse = mean_squared_error(y_test, y_predict)
rmse = math.sqrt(mse)
print(mse)
print(rmse)
```
