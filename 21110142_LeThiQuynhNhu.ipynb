Lab 4: Input: vector dữ liệu một chiều

loại kiểm định (lớn, bé, bằng)

Độ tin cậy alpha = 0.95

Phương sai

Output: loại kiểm định

p-value

Bậc tự do

Khoảng tin cậy

```python
import math
import scipy.stats as stats

def one_sample_t_test(data, mu0, test_type, alpha):
    # Tính x trung bình
    x_bar = sum(data) / len(data)

    # Tính độ lệch chuẩn
    n = len(data)
    sample_std = math.sqrt(sum((x - x_bar) ** 2 for x in data) / (n - 1))
    sample_std_square = sum((x - x_bar) ** 2 for x in data) / (n - 1)
    print('Phương sai mẫu: ', sample_std_square)

    # Tính chi-square
    chi_square = (n-1)*sample_std_square/19
    print(chi_square)
    

    # tính bậc tự do
    df = n - 1
    print('bậc tự do: ', df)

    # Tính p-value
    if test_type == 'greater':
        p_value = 1 - stats.chi2.cdf(chi_square, df=n - 1)
    elif test_type == 'less':
        p_value = stats.chi2.cdf(chi_square, df=n - 1)
    elif test_type == 'equal':
        p_value = 2 * min((1 - stats.chi2.cdf(abs(chi_square), df=n - 1)))

    # Xác định loại kiểm định
    if p_value < alpha:
        test_result = 'Reject H0'
    else:
        test_result = 'Fail to reject H0'

    # Tính khoảng tin cậy
    t_critical = stats.chi2.ppf(1 - alpha/2, df)
    margin_of_error = (n-1)*sample_std_square/t_critical
    confidence_interval = (x_bar - margin_of_error, x_bar + margin_of_error)

    return x_bar, p_value, test_result, confidence_interval

# Đầu vào - Mảng dữ liệu của bạn, giá trị μ0, loại kiểm định ('greater', 'less', 'equal'), và mức ý nghĩa alpha
data = [129, 132, 140, 141, 138, 143, 133, 137, 140, 143, 138, 140]

mu0 = 90  # Giá trị μ0
test_type = 'greater'  # Loại kiểm định ('greater', 'less', 'equal')
alpha = 0.05  # Mức ý nghĩa

# Thực hiện kiểm định thống kê và tính các giá trị
x_bar, p_value, test_result, confidence_interval = one_sample_t_test(data, mu0, test_type, alpha)

# In kết quả
print(f"Giá trị trung bình của mẫu: {x_bar}")
print(f"P-value: {p_value}")
print(f"Kết quả kiểm định: {test_result}")
print(f"Khoảng tin cậy: {confidence_interval}")

```

Lab5: Viết hàm kiểm định hai mẫu cho trung bình

Input: 2 vector dữ liệu

Độ tin cậy

Phương sai 1, Phương sai 2, nếu có độ lệch chuẩn tổng thể

Output: Phương sai mẫu 1,2

Trung bình mẫu 1,2

p-value

mức ý nghĩa



```python
import scipy.stats as stats

def two_sample_t_test(data1, data2, alpha=0.05):
    # Số lượng mẫu và trung bình mẫu
    n1 = len(data1)
    n2 = len(data2)
    mean1 = sum(data1) / n1
    mean2 = sum(data2) / n2
    
    # Phương sai mẫu
    var1 = sum((x - mean1)**2 for x in data1) / (n1 - 1)
    var2 = sum((x - mean2)**2 for x in data2) / (n2 - 1)
    
    # Kiểm định t-Student
    t_statistic, p_value = stats.ttest_ind(data1, data2, equal_var=False)
    
    # Xác định mức ý nghĩa
    if p_value < alpha:
        significance = "Có ý nghĩa"  # Từ chối giả thuyết không có sự khác biệt
    else:
        significance = "Không có ý nghĩa"  # Không đủ bằng chứng để từ chối giả thuyết
    
    return var1, var2, mean1, mean2, p_value, significance

# Đầu vào - Hai vector dữ liệu
data1 = [0.97, 0.72, 1.00, 0.81, 0.62, 1.32, 1.24, 0.99, 0.9, 0.74, 0.88, 0.94, 1.16, 0.86, 0.85, 0.58, 0.57, 0.64, 0.98, 1.09, 0.92, 0.78, 1.24, 1.18]
data2 = [0.48, 0.71, 0.98, 0.68, 0.68, 1.18, 1.36, 0.78, 1.64]

alpha = 0.05  # Độ tin cậy

# Thực hiện kiểm định và tính toán các giá trị liên quan
var1, var2, mean1, mean2, p_value, significance = two_sample_t_test(data1, data2, alpha)

# In kết quả
print(f"Phương sai mẫu 1: {var1}")
print(f"Phương sai mẫu 2: {var2}")
print(f"Trung bình mẫu 1: {mean1}")
print(f"Trung bình mẫu 2: {mean2}")
print(f"P-value: {p_value}")
print(f"Mức ý nghĩa: {significance}")
```
